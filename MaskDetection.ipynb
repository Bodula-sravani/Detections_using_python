```python
from keras.models import load_model
#from keras.preprocessing.image import img_to_array
from tensorflow.keras.utils import img_to_array
from keras.preprocessing import image
import cv2
import numpy as np
```

```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
#face_classifier=cv2.CascadeClassifier('D:\\a\\opencv-python\\opencv-python\\opencv\\build\\etc\\haarcascades\\haarcascade_frontalface_default.xml')
from tensorflow.keras.models import model_from_json
classifier = model_from_json(open("mask_model.json", "r").read())
classifier.load_weights('mask_model.h5')
from tensorflow.keras.models import model_from_json
#test = face_cascade.load('haarcascade_frontalface_default.xml')
```

```python
results={0:'NO MASK',1:'MASKED'}
GR_dict={0:(0,0,255),1:(0,255,0)}
rect_size = 4
cap=cv2.VideoCapture(0)
font = cv2.FONT_HERSHEY_SIMPLEX
while True:
    
    (rval, im) = cap.read()
    im=cv2.flip(im,1,1) 
    
    rerect_size = cv2.resize(im, (im.shape[1] // rect_size, im.shape[0] // rect_size))
    faces = face_cascade.detectMultiScale(rerect_size)
    for f in faces:
        (x, y, w, h) = [v * rect_size for v in f] 
        
        face_img = im[y:y+h, x:x+w]
        rerect_sized=cv2.resize(face_img,(150,150))
        normalized=rerect_sized/255.0
        reshaped=np.reshape(normalized,(1,150,150,3))
        reshaped = np.vstack([reshaped])
        result=classifier.predict(reshaped)
        
        label=np.argmax(result,axis=1)[0]
      
        cv2.rectangle(im,(x,y),(x+w,y+h),GR_dict[label],2)
        cv2.rectangle(im,(x,y-40),(x+w,y),GR_dict[label],-1)
        cv2.putText(im, results[label], (x, y-10),cv2.FONT_HERSHEY_SIMPLEX,0.8,(255,255,255),2)
    
    #cv2.namedWindow("face mask", cv2.WINDOW_NORMAL)
    #cv2.resizeWindow("mask", 1500, 1500)
    cv2.imshow('face mask',im)
    if cv2.waitKey(1) & 0xFF == ord('e'):
        break
cap.release()
cv2.destroyAllWindows()
```
